```python
import os
import pandas as pd

# Define paths
input_folder = 'CSV files IDX'
output_listings_path = 'combined_listings_residential.csv'
output_sold_path = 'combined_sold_residential.csv'

# Initialize lists to hold dataframes
listings_frames = []
sold_frames = []

# Counters for required row-count tracking
raw_listings_total = 0
raw_sold_total = 0

# Set up loop parameters from January 2024 to May 2026
year = 2024
month = 1

print("🔄 Beginning data compilation with suffix matching...")

while (year < 2026) or (year == 2026 and month <= 5):
    suffix = f"{year}{month:02d}"
    
    # -----------------------------------------------------------------
    # Process Monthly Listing File (Checking both normal and _filled)
    # -----------------------------------------------------------------
    listing_file = os.path.join(input_folder, f"CRMLSListing{suffix}.csv")
    listing_file_filled = os.path.join(input_folder, f"CRMLSListing{suffix}_filled.csv")
    
    # Determine which file version actually exists
    chosen_listing = listing_file_filled if os.path.exists(listing_file_filled) else listing_file
    
    if os.path.exists(chosen_listing):
        df_list = pd.read_csv(chosen_listing)
        raw_listings_total += len(df_list)
        listings_frames.append(df_list)
    else:
        print(f"⚠️ Warning: Missing file {listing_file} (or _filled version)")
        
    # -----------------------------------------------------------------
    # Process Monthly Sold File (Checking both normal and _filled)
    # -----------------------------------------------------------------
    sold_file = os.path.join(input_folder, f"CRMLSSold{suffix}.csv")
    sold_file_filled = os.path.join(input_folder, f"CRMLSSold{suffix}_filled.csv")
    
    # Determine which file version actually exists
    chosen_sold = sold_file_filled if os.path.exists(sold_file_filled) else sold_file
    
    if os.path.exists(chosen_sold):
        df_sold = pd.read_csv(chosen_sold)
        raw_sold_total += len(df_sold)
        sold_frames.append(df_sold)
    else:
        print(f"⚠️ Warning: Missing file {sold_file} (or _filled version)")

    # Move to next calendar month
    month += 1
    if month > 12:
        month = 1
        year += 1

# =====================================================================
# CONCATENATION AND FILTERING
# =====================================================================

# 1. Concatenate all frames together
combined_listings_raw = pd.concat(listings_frames, ignore_index=True) if listings_frames else pd.DataFrame()
combined_sold_raw = pd.concat(sold_frames, ignore_index=True) if sold_frames else pd.DataFrame()

# Verify initial concatenation counts
concat_listings_count = len(combined_listings_raw)
concat_sold_count = len(combined_sold_raw)

# 2. Apply the 'Residential' PropertyType filter
combined_listings_filtered = combined_listings_raw[combined_listings_raw['PropertyType'] == 'Residential'] if not combined_listings_raw.empty else pd.DataFrame()
combined_sold_filtered = combined_sold_raw[combined_sold_raw['PropertyType'] == 'Residential'] if not combined_sold_raw.empty else pd.DataFrame()

final_listings_count = len(combined_listings_filtered)
final_sold_count = len(combined_sold_filtered)

# Save the final dataframes to new CSV files
combined_listings_filtered.to_csv(output_listings_path, index=False)
combined_sold_filtered.to_csv(output_sold_path, index=False)

print("\n🎉 Process complete! Output files generated successfully.")

# =====================================================================
# REQUIRED DELIVERABLE COMMENTS (ROW COUNT CONFIRMATIONS)
# =====================================================================
# Total individual row counts compiled across all raw files:
# Raw Listings Cumulative Sum: {raw_listings_total}
# Raw Sold Cumulative Sum: {raw_sold_total}
#
# ROW COUNTS BEFORE THE RESIDENTIAL FILTER (Post-Concatenation):
# - Combined Listings Count: {concat_listings_count} rows
# - Combined Sold Count: {concat_sold_count} rows
#
# ROW COUNTS AFTER THE RESIDENTIAL FILTER (Final Output):
# - Filtered Residential Listings: {final_listings_count} rows
# - Filtered Residential Sold: {final_sold_count} rows
# =====================================================================

# Print metrics to terminal
print(f"\n📈 Final Row Verification Metrics:")
print(f"   ↳ Listings: {concat_listings_count} rows concatenated -> {final_listings_count} after Residential filter.")
print(f"   ↳ Sold: {concat_sold_count} rows concatenated -> {final_sold_count} after Residential filter.")
```
